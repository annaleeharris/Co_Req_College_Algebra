#### Plot example with title, labels, fontsize, other attributes

```julia
# Graph y = 3x with a title and label

x = 1:10      # x = {1,2,3,...,10}
y = 3*x       # y = 3x
plot(x,y, title = "Graph of y = 3x", label = "Line", lw = 10,)
```

```julia
# Graph y = x^2

x = -10:0.1:10            # x = {-10, -9,...8,9,10}
y = x.^2              # y = x^2; Remember to put "." after x so it can numerically compute.
```

```julia
(collect(x),  collect(y))
```

```julia
plot(x,y)
```

```julia
# More examples of plots

x = 1:10           # x = {1,2,...,9,10}
y = rand(10)       # y = randomly generated 10 numbers
plot(x,y)          # Graph a line graph
```

```julia
# Graph y = 3x^4 + 2x^3 - 4x^2 + 5x + 7

x = 1:10                                      # x = {1,2,...,9,10}
y = x.^4 .+ 2*x.^3 .- 4*x.^2 + 5*x .+ 7          #some polynomial
plot(x,y)                                     # graph the polynomial
```
